```python
from langchain.document_loaders import UnstructuredURLLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import Pinecone
import pinecone
from langchain.embeddings import HuggingFaceEmbeddings
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
import transformers
from langchain import HuggingFacePipeline
from huggingface_hub import  notebook_login
import torch
```

```python
urls = ["https://www.cuchd.in/"]
```

```python
# loading the data from the url
```

```python
loader = UnstructuredURLLoader(urls = urls)
data = loader.load()
data
```

```python
# chunking
```

```python
text_splitter = CharacterTextSplitter(separator= '\n', chunk_size=1000, chunk_overlap=200)
```

```python
text_chunks = text_splitter.split_documents(data)
```

```python
len(text_chunks)
```

```python
# embedding
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
query = "chatbot!"
response = embeddings.embed_query(query)
len(response)
```

```python
import pinecone
# initialising the pinecone
pinecone.init(
    api_key = PINECONE_API_KEY,
    environment = PINECONE_API_ENV
)

index_name = "chatbot"
```

```python
docsearch = Pinecone.from_texts([t.page_content for t in text_chunks], embeddings, index_name = index_name)
```

```python
# create llm wrapper
```

```python
model_name = "meta-llama/Llama-2-7b-chat-hf"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name ,token= True)
```

```python
token = "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"
```

```python

from bitsandbytes.optim import Adam8bit
```

```python
quantization_config = BitsAndBytesConfig(use_8bit=True)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map='auto', torch_dtype=torch.float16,token= token, quantization_config=quantization_config)
```

```python
pipe = transformers.pipeline(
    "text-generation",
    model=model_name,  # Replace "model_name" with the actual model name or identifier
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    device_map='auto',
    max_new_tokens=256,
    do_sample=True,
    top_k=20,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    batch_size=2,

)
```

```python
pipe.tokenizer.pad_token_id = model.config.eos_token_id
```

```python
llm = HuggingFacePipeline(pipeline = pipe , model_kwargs={"temperature":0})
```

```python
# now we will ask the question from the website which we passsed
# so we will use retriecal qa
```

```python
from langchain.chains import RetrievalQA
```

```python
qa = RetrievalQA.from_chain_type(llm = llm, chain_type="stuff", retriever= docsearch.as_retriever())
```
